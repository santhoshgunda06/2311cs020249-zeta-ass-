```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Day_12_banking_data.csv')
df
```

```python
total_transaction_amount_by_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()
total_transaction_amount_by_account_type.plot(kind='bar')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.show()
```
